```
#Importing the required libraries
import numpy as np
import pandas as pd
```

```
#Loading the dataset to be used in the task
df = pd.read_csv('http://bit.ly/autolib_dataset')
df.head()
```

```
#Check for the information in the dataset
df.info()
```

```
#Cleaning the dataset by dropping the cars column since its a duplicate of the bluecars column
df1 = df.drop(['Charge Slots','Charging Status','Rental status','Slots','year','month','minute','Cars','Displayed comment','Geo point','Scheduled at','ID'],axis=1)
df1.head()
```

```
#Checking for any null values in the dataset
df1.isnull().sum()
```

```
#Converting column names to lowercase
df1.columns = df1.columns.str.lower()
df1.head()
```

```
#Most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
PopularH = df1[df1['city'] == 'Paris']
CorrectH = PopularH.groupby(['public name', 'hour'])[['bluecar counter']].sum()
CorrectH.head()
```

```
CorrectH['bluecar_usage'] = CorrectH['bluecar counter'].diff()

CorrectH.head()
```
